```python
first= text
second= text2
#split so each word have their own string
first = first.split(" ")
second= second.split(" ")
#print(first,second)
total= set(first).union(set(second))
#print(total)
wordDictA = dict.fromkeys(total, 0) 
wordDictB = dict.fromkeys(total, 0)
#worddictA => all keys but val !=0 only for words in those docs 
for word in first:
    wordDictA[word]+=1
    
for word in second:
    wordDictB[word]+=1
#put them in a dataframe and then view the result:
pd.DataFrame([wordDictA, wordDictB])
```

```python
def computeTF(wordDict,bow):
    tfDict = {}
    bowCount = len(bow)
    for word,count in wordDict.items():
        tfDict[word] = count/float(bowCount)
    return tfDict
```

```python
def computeIDF(docList):
    idfDict = {}
    N = len(docList)
    
    idfDict = dict.fromkeys(docList[0].keys(), 0.0)
    for doc in docList:
        for word, val in doc.items():
            print(doc, idfDict[word])
            if val > 0:
                idfDict[word] += 1.0
    
    for word, val in idfDict.items():
        idfDict[word] = math.log10(N / float(val)) # N = 2, float val 1 or 2 => log (2/1) or log(2/2) => 0.3010 or 0
        
    return idfDict
#inputing our sentences in the log file
idfs = computeIDF([wordDictA, wordDictB])
```
