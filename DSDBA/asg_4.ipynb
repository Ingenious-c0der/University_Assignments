```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("titanic.csv")
df.shape
```

```python
df.describe()
```

```python
sns.barplot(x = df['Sex'],y = df['Survived'])
```

```python
sns.barplot(x = df['Pclass'],y = df['Survived'])
```

```python

df['MartialStatus'] = 1

for index , row in df.iterrows():
    name = row["Name"]
    if "Miss." in name:
         df.loc[index,"MartialStatus"] = 0
    elif "Mr." in name:
         df.loc[index,"MartialStatus"] = 2

    elif "Master." in name and row.Age < 18 :
         df.loc[index,"MartialStatus"] = 3
    
    #df.loc[index,"MartialStatus"] = 0 if "Miss." in row['Name'] else 1 if "Mr." in row['Name'] else 2
sns.barplot(x = df['MartialStatus'], y = df['Survived'])
```

```python
sns.barplot(x = df.SibSp, y = df.Survived)
```

```python

sns.set(rc={'figure.figsize':(50,50)})
x = sns.histplot(data = df,x= df.Fare,kde=True,hue = "Pclass")
x.set_xlim(0,100)
```

```python
# Create the box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Sex', y='Age', hue='Survived')
plt.title('Distribution of Age by Gender and Survival Status')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.legend(title='Survived', loc='upper right')
plt.show()
```
