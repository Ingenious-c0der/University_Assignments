```python
df = pd.read_csv('titanic.csv')

# Filter out missing values for the 'Age' column
df_age = df.dropna(subset=['Age'])

# Create the violin plot
plt.figure(figsize=(8, 6))
sns.violinplot(x='Survived', y='Age', data=df_age, palette='Set3')

# Customize the plot
plt.title('Violin Plot of Age by Survival')
plt.xlabel('Survived')
plt.ylabel('Age')

# Show the plot
plt.show()
```
