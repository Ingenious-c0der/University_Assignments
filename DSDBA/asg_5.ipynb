```python
import pandas as pd 
import numpy as np
from statsmodels.graphics.mosaicplot import mosaic
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("titanic.csv")
```

```python
plt.rc('figure', figsize=(17, 5))
```

```python
mosaic(df,['Survived', 'Pclass', 'Sex'], title='Survival: Red=Died, Green=Survived')
```

```python
age_sex_hist = sns.FacetGrid(df, col='Survived', row='Sex', hue='Survived')
age_sex_hist.map(plt.hist, 'Age', bins=20)
```

```python
sns.boxplot(df, x = "Pclass",y="Age", hue = "Survived")                   
```
